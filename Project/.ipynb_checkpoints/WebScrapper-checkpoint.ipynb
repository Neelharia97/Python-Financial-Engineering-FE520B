```python
import requests
from bs4 import BeautifulSoup
import urllib
import pandas as pd
```

```python
def generate_url(base_url, component):
    url = base_url
    for r in component:
        url = '{}/{}'.format(url,r)
        
    return url

base_url = r"https://www.sec.gov/Archives/edgar/data"
component = ['898293','00012252082001396','0001225208-20-013967.html']
test_url = generate_url(base_url,component)


new_base_url = r"https://www.sec.gov/Archives/edgar/daily-index"
year_url = generate_url(new_base_url,['2019','index.json'])
print(year_url)
content = requests.get(year_url)
decoded = content.json()
#decoded
```

```python
for item in decoded['directory']['item']:
    print('-'*100)
    print('Url for quarter {}'.format(item['name']))
    qtr_url =  generate_url(new_base_url,['2019',item['name'],'index.json'])
    print(qtr_url)
    
    file_content = requests.get(qtr_url)
    decoded = file_content.json()
    l = []
           
    for file in decoded['directory']['item']:
        file_url = generate_url(new_base_url,['2019',item['name'],file['name']])
        print(file_url)
        l.append(file_url)
  
    
file_url = r"https://www.sec.gov/Archives/edgar/daily-index/2019/QTR2/master.20190401.idx"         
file_content = requests.get(file_url).content
#file_name = 'master_{}'.format(file_url[-4])
with open('master_2019','wb') as f:
    f.write(file_content)
          
```

```python
with open('master_2019','rb') as f:
    byte_data = f.read()

data = byte_data.decode('UTF-8').split(' ')
for index,item in enumerate(data):
    if 'ftp://ftp.sec.gov/edgar/' in item:
        start_index = index 
        print(start_index)
master_list= []

format_data = data[start_index+1:]
format_data
```

```python

master_list = []
for index, item in enumerate(format_data):
    if index == 0:
        clean_data = item.replace("\n",'|').split('|')
        clean_data = clean_data[8:]
    else:
        clean_data = item.replace("\n",'|').split('|')
   

    for index, row in enumerate(clean_data):
        if '.txt' in row:
            mini_list = clean_data[(index-4):index+1]
            
            if len(mini_list) != 0:
                mini_list[4] = "https://www.sec.gov/Archives/"+mini_list[4]
                master_list.append(mini_list)
master_list[:3]

```

```python
print(len(master_list))
for index, row in enumerate(master_list):
    dict1 = {
        'CIK': row[0],
        'Company Name' : row[1],
        'Form Type' : row[2],
        'Date Filed' :row[3],
        'File Nane' : row[4]
    }
    master_list[index] = dict1
```
